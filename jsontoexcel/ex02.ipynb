```python
import pandas as pd
import re

# 엑셀 파일에서 데이터 읽기
df = pd.read_excel('data.xlsx', header=None)

# 엑셀 파일에서 읽은 데이터를 문자열로 변환 후 하나의 텍스트로 결합
data = " ".join(df[0].astype(str).tolist())

# 공백을 하나로 바꾸고, 불필요한 줄바꿈 제거
cleaned_data = re.sub(r'\s+', ' ', data).strip()

# 데이터의 각 항목을 쉼표로 나누어 리스트로 추출
lines = cleaned_data.split(", ")

# 각 항목을 ':' 기준으로 나누어 키와 값으로 분리
split_lines = []
for line in lines:
    parts = line.split(":", 1)  # 최대 한 번만 나누기
    if len(parts) == 2:
        split_lines.append(parts)
    else:
        split_lines.append([parts[0], ""])  # 값이 없는 경우 처리

# 분리된 데이터로 데이터프레임 만들기
result_df = pd.DataFrame(split_lines, columns=["Key", "Value"])

# 엑셀 파일로 저장
result_df.to_excel('super_hero_squad.xlsx', index=False)

print("엑셀 파일로 저장되었습니다.")
```
